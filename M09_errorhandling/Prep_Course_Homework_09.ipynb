1) Con la clase creada en el módulo 7, tener en cuenta diferentes casos en que el código pudiera arrojar error. Por ejemplo, en la creación del objeto recibimos una lista de números enteros pero ¿qué pasa si se envía otro tipo de dato?

```python
class funcionesUtiles:

    def __init__(self):
        pass

    def definiPrimo(self, numero):
        try:
            numero = abs(int(numero))
            if (numero == 0 or numero == 1):
                return False
            elif (numero == 2):
                return True
            else:
                for i in range (2, numero):
                    if (numero % i == 0):
                        return False
                return True
        except ValueError:
            return "Error: El valor ingresado no es un número entero válido."

    def conversionTemperatura(self, valor, medidaOrigen, medidaDestino):
        try:
            if not isinstance(valor, (int, float)):
                raise TypeError("El valor debe ser un número.")
            if medidaOrigen not in ['Celsius', 'Fahrenheit', 'Kelvin']:
                raise ValueError("La medida de origen debe ser 'Celsius', 'Fahrenheit' o 'Kelvin'.")
            if medidaOrigen == 'Celsius':
                if medidaDestino == 'Fahrenheit':
                    return (valor * 9/5) + 32
                elif medidaDestino == 'Kelvin':
                    return valor + 273.15
            elif medidaOrigen == 'Fahrenheit':
                if medidaDestino == 'Celsius':
                    return (valor - 32) * 5/9
                elif medidaDestino == 'Kelvin':
                    return (valor - 32) * 5/9 + 273.15
            elif medidaOrigen == 'Kelvin':
                if medidaDestino == 'Celsius':
                    return valor - 273.15
                elif medidaDestino == 'Fahrenheit':
                    return (valor - 273.15) * 9/5 + 32
        except TypeError:  
            return "Error: El valor ingresado no es un número válido."
        except ValueError as e:
            return f"Error: {str(e)}"

    def factorial(self, numero):
            try:
                if not isinstance(numero, int) or isinstance(numero, float) or isinstance(numero, bool):
                    raise ValueError("El valor debe ser un número entero.")
                numero = abs(int(numero))  
                if numero == 0:
                    return 1
                elif numero > 1:
                    return numero * self.factorial(numero - 1)
                else:
                    return numero
            except ValueError:
                return "Error: El valor ingresado no es un número entero válido."
    
numero = funcionesUtiles()    
print("El factorial de 5 es: ", numero.factorial(True))
print("Conversión de 100 Celsius a Fahrenheit:", numero.conversionTemperatura(100, 'C', 'F'))
```
