6) Utilizar un ciclo while para realizar el factorial de un número guardado en una variable, sólo si la variable contiene un número entero mayor a 0

```python
numeroFactorial = int(3)
def factorial(numeroFactorial):
    if numeroFactorial == 0:
        return 1
    else:
        print("El numero es: " + str(numeroFactorial))
        temp = int(numeroFactorial * factorial(numeroFactorial - 1))
        print(temp)
        return temp
    
print("El factorial de " + str(numeroFactorial) + " es: " + str(factorial(numeroFactorial)))
```

9) Imprimir los números primos existentes entre 0 y 30

```python
numeroPrimo = int(31)
banderaPrimo  = bool(False)

for i in range(0, numeroPrimo):
    for j in range(0 , i):
        ##print("i "+ str(i) + " j " + str(j))
        if((i == 0) or (j == 0) or (i == 1) or (j == 1)):
            continue
        elif(i % j == 0):
            print("Numero " + str(i) + " no es primo, el divisor es " + str(j))
            banderaPrimo = True
            break;

    if(banderaPrimo == False):
        print("Numero " + str(i) + " es primo")
    else:
        banderaPrimo = False
```

12) Aplicando continue, armar un ciclo while que solo imprima los valores divisibles por 12, dentro del rango de números de 100 a 300

```python
rangoInferior = int(100)
rangoSuperior = int(300)
contador = int(rangoInferior)
while (contador >= rangoInferior and contador <= rangoSuperior):
    while((contador % 12) != 0):
        print("contador " + str(contador))
        contador = contador + 1
        continue
    print("El numero " + str(contador) + " es divisible por 12")
    contador = contador + 1
```

13) Utilizar la función **input()** que permite hacer ingresos por teclado, para encontrar números primos y dar la opción al usario de buscar el siguiente

```python
numeroEntero = int(input("Ingrese un numero entero para saber si es primo o no: "))

if numeroEntero < 2:
    print("El numero " + str(numeroEntero) + " no es primo")
else:
    for i in range(2, int(numeroEntero / 2) + 1):
        if (numeroEntero % i) == 0:
            print("El numero " + str(numeroEntero) + " no es primo")
            break
    else:
        print("El numero " + str(numeroEntero) + " es primo")
```
